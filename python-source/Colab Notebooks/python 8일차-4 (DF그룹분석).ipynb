```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
tips=sns.load_dataset('tips')
```

* 각 품종별 features

```python
iris.groupby('species').mean()
```

```python
iris.groupby('species').first()   #그룹 내 non-NA 값 중 첫번째 값
```

* 각 품종별 sepal_length의 표준편차

```python
iris.groupby('species').std()['sepal_length']
```

```python
iris[['sepal_length','species']].groupby('species').std()
```

* 그룹 연산을 여러가지 할 경우
* 각 품종별 sepal_length의 평균,표준편차,최대값,최소값

```python
iris.groupby('species')['sepal_length'].agg(['mean','std','min','max']) #agg=aggregate 
```

#2. Pivot table

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}

df=pd.DataFrame(data)
df.head()
```

```python
df.pivot_table('인구','도시','연도')
```

```python
df.pivot_table('인구','도시','연도',margins=True,margins_name="평균")
```

```python
df.pivot_table('인구',index=['연도','도시'])
```

```python
tips.groupby(['sex','smoker']).size()
```

```python
tips.pivot_table('tip_pct','sex','smoker',aggfunc='count',margins=True,margins_name='계')
```

```python
tips.groupby('sex')[['tip_pct']].mean()
```

```python
tips.pivot_table('tip_pct','sex')
```

```python
# 흡연 유무에 따른 평균 팁 비율과 최대 최소
tips.groupby('smoker')['tip_pct'].agg(['mean','max','min'])
```

```python
tips.pivot_table('tip_pct','smoker',aggfunc=['mean','max','min'])
```

```python
tips.pivot_table('tip_pct','sex','smoker',aggfunc='count',margins=True,margins_name='계')
```

```python
#성별,흡연유무에 따른 평균 팁 비율
tips.pivot_table('tip_pct',['sex','smoker'])
```

```python
tips.pivot_table('tip_pct','sex','smoker')
```
