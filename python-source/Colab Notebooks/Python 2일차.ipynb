```python
if mine==yours:
  result=draw
else:
  #print('잘모름')
  if mine==scissor:
    if yours==rock:
      result=lose
    else:
      result=win
  else:
    if mine==rock:
      if yours==paper:
        result=lose
      else:
        result=win
    else:
      if mine==paper:
        if yours==scissor:
          result=lose
        else:
          result=win
      else:
        print('이상해요')
print(result)
```

```python
string="""나는 홍길동입니다
주소는 서울시입니다
나이는 35세입니다"""

sents=[]
words=[]

for sen in string.split(sep="\n"):
  sents.append(sen)
  for word in sen.split():
    words.append(word)
print("문장:",sents)
print("문장수:",len(sents))
print("단어:",words)
print("단어수:",len(words))
```
