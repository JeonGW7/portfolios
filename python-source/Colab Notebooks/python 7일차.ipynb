* 결측치(NA,NaN,Null)처리 
isna,fillna,dropna

```python
dm=metro-metro2
dm
```

```python
dm.isna()  #결측치가 있는지 확인
```

```python
dm.isnull()  #isna와 같음
```

```python
dm.isna().sum(),dm.isnull().sum() #결측치의 갯수
```

```python
dm.fillna(dm.mean())  #평균값으로 채움
```

```python
dm.fillna(dm.median())  #중앙값으로 채움
```

```python
#결측치를 0으로 대체하고 정수형으로 변경
dm.fillna(0).astype(int)
```

```python
dm.dropna()
```

inplace option

```python
dm2=dm.dropna()  #값이 null이 들어있으면 삭제
dm2
```

```python
dm.dropna(inplace=True)
dm
```
