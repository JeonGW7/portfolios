* B   business day frequency 평일
* C   custom business day frequency
* D   calendar day frequency
* W   weekly frequency 주말 
* W-SUN 일 / 토 W-SAT / 월 W-MON
* M   month end frequency 매월의 마지막 날짜 출력
* SM   semi-month end frequency (15th and end of month)
* BM   business month end frequency
* CBM   custom business month end frequency
* MS   month start frequency 매월의 첫날짜 출력
* SMS   semi-month start frequency (1st and 15th)
* BMS   business month start frequency 매월 첫 영업일을 출력 
* CBMS   custom business month start frequency
* Q   quarter end frequency

```python
pd.date_range('20220401',periods=30,freq='B')  #평일만 출력
```

```python
pd.date_range('20220401','20220430',freq='W')  #주말만 출력
```

* resample

```python
#periods 매개변수는 date_range함수에 전달된 시작날짜와 종료날짜를 periods
#매개변수에 전달된 기간만큼 동일하게 나누어 출력해준다
np.random.seed(2022)
ts=pd.Series(np.random.randint(100,1000,100),index=pd.date_range('20220401',periods=100))
ts
```

```python
ts.resample('M').first()  #매월의 마지막일 출력
```
