```python
import folium
```

```python
map=folium.Map(location=[37.531, 126.986],zoom_start=12)
map
```

```python
folium.Marker([37.566317, 126.977829], popup=folium.Popup('서울시 중구 세종대로 110',max_width=200),tooltip='서울특별시청').add_to(map)
map
```

```python
#지도위에 서울시청,서초구청,강남구청
#tiles:Stamen Terrain, Stamen Toner, Open Street Map(default)
map=folium.Map(location=[37.531, 126.986],zoom_start=12,tiles='Stamen Toner')
folium.Marker([37.566317, 126.977829], popup=folium.Popup('서울시 중구 세종대로 110',max_width=200),tooltip='서울특별시청').add_to(map)
folium.Marker([37.484731, 127.032078], popup=folium.Popup('서울시 서초구 남부순환로 2584',max_width=200),tooltip='서초구청',icon=folium.Icon(color='red',icon='info-sign')).add_to(map)
folium.Circle([37.517236, 127.047324],tooltip='강남구청',fill=True,color='crimson',radius=300).add_to(map)
title_html='<h3 align="center" style="font-size:20px">3개지역표시</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
