* Alice in wonderland

```python
with open('Alice.txt') as file:
  text=file.read()
```

```python
# stopwords 추가
sw_wc.add('said')
```

```python
wc=WordCloud(background_color='white',max_words=2000,stopwords=sw_wc)
wc=wc.generate(text)
```

```python
count=0
for key,val in wc.words_.items():
  print(key,':',val)
  count+=1
  if count > 10: 
    break
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wc,interpolation='bilinear')  
plt.axis('off')
plt.show()
```

* 그림 위에 표시하기

```python
mask=np.array(Image.open('Alice.png'))
plt.imshow(mask,cmap=plt.cm.gray,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc=WordCloud(background_color='white',max_words=2000,stopwords=sw_wc,mask=mask).generate(text)
plt.figure(figsize=(10,12))
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
with open('A_new_hope.txt') as file:
  text=file.read()
```

```python
text= text.replace('HAN','Han').replace("LUKE's","Luke")
```

```python
sw=STOPWORDS
sw.add('int')
sw.add('ext')
len(sw)
```

```python
mask=np.array(Image.open('A_new_hope.png'))
plt.imshow(mask,cmap=plt.cm.gray,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc=WordCloud(background_color='white',max_words=1000,stopwords=sw,mask=mask,margin=10,random_state=2022).generate(text)
default_colors=wc.to_array
plt.figure(figsize=(10,12))
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc=WordCloud(max_words=1000,stopwords=sw,mask=mask,margin=10,random_state=2022).generate(text)
default_colors=wc.to_array
```

```python
def grey_color(word,font_size,position,orientation,random_state,**kwargs):
  return f'hsl(0,0%,{np.random.randint(60,101)}%)'
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=grey_color,random_state=2022),interpolation='bilinear')
plt.axis('off')
plt.show()
```
