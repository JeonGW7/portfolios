대각선:identity,eye

```python
#정방향 (n by n)
np.identity(3,dtype=np.int32) #대각선 방향으로 1 
```

```python
#직사각형 모양 (n by m)
np.eye(N=3,M=5,dtype=int)
```

```python
from types import DynamicClassAttribute
np.eye(3,5,k=2,dtype=int) #k의 값에 따라 1이 들어가는 위치가 바뀜 
```

```python
np.eye(3,5,k=1,dtype=int)
```

```python
np.eye(3,5,k=-1,dtype=int)
```

```python
np.eye(3,5,k=0,dtype=int) +np.eye(3,5,k=2,dtype=int)
```

random sampling

```python
#데이터 분포에 따른 sampling 배열을 생성 
np.random.seed(2022)  #난수를 예측 가능하게 해준다 시드값이 똑같으면 결과값이 동일하게 나온다
np.random.randint(1,7,10)   #1~7사이 10개
```

```python
#균등분포(Uniform) (0~1사이 실수값)
np.random.random(4)
```

```python
np.random.uniform(3,4,4)  #3-4사이의 실수값 
```

```python
#표준정규분포 N(0,1)
np.random.randn(4)
```

```python
#이항분포
np.random.binomial(10,0.5,12).reshape(3,4)  #50퍼확률 12번 
```

```python
#t분포 : 모집단 표준편차를 알 수 없을 때 표본 평균과 모집단 평균 사이 표준화된 거리를 설명하며, 관측값은 정규 분포를 따르는 모집단에서 추출된다
np.random.standard_t(2,4)  #자유도2, 4개 갯수
```
