#Matplotlib 다양한 플롯

1.산점도(Scatter plot)

* x와 y의 관계를 눈으로 확인하고자 할 때

```python
X,Y= np.random.randn(100),np.random.randn(100)
plt.title('산점도')
plt.scatter(X,Y)
plt.show()
```

```python
color=np.random.rand(100)
plt.scatter(X,Y,c=color,alpha=0.5)
plt.title('산점도')
plt.show()
```

```python
area=500*np.random.rand(100)
plt.scatter(X,Y,c=color,alpha=0.5,s=area)
plt.title('차트')
plt.show()
```

2. Barchart

```python
import seaborn as sns
titanic=sns.load_dataset('titanic')
titanic['class'].value_counts()
```

```python
s=titanic['class'].value_counts()
plt.barh(s.index,s.values)  
plt.title('타이타닉의 선실별 승객수')
plt.show()
```

3. Histogram

```python
s=np.random.rand(200)
plt.hist(s)
plt.title('Histogram')
plt.show()
```

```python
s=np.random.rand(2000)
plt.hist(s,bins=20)
plt.title('Histogram')
plt.show()
```

4. boxplot

```python
iris=sns.load_dataset('iris')
```

```python
setosa=iris[iris.species=='setosa']
setosa.head()
```

```python
plt.boxplot(setosa.sepal_length)
plt.title('Seotsa 품종 sepal_length')
plt.show()
```

```python
plt.boxplot(setosa.sepal_width)
plt.title('Seotsa 품종 sepal_width')
plt.show()
```

```python
plt.boxplot([setosa.sepal_width,setosa.sepal_length],labels=iris.columns[:2])
plt.title('Seotsa 품종 꽃받침의 길이와 너비')
plt.show()
```

* 파이(pie)차트

```python
ratio=[10,15,20,25]
labels=['감','귤','배','밤']
plt.pie(ratio,labels=labels,autopct="%.1f%%")
plt.title('파이차트')
plt.show()
```

```python
colors=['yellowgreen','gold','lightskyblue','lightcoral']
explode=(0,0,0.4,0)
plt.pie(ratio,explode=explode,colors=colors,labels=labels,autopct="%.1f%%")
plt.title('pie chart')
plt.show()
```
