```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
titanic=sns.load_dataset('titanic')
titanic
```

```python
titanic['age_grade']=pd.qcut(titanic['age'],3,labels=["청년","중년","장년"])
```

```python
titanic.pivot_table( index = ['age_grade', 'sex'], columns = 'class', values = 'survived',aggfunc = lambda x : (sum(titanic['survived']) / sum(x)))
```

```python
#다른방법
titanic['category1']=titanic.apply(lambda r : r.sex if r.age >20 else 'child', axis=1)
```

```python
titanic.age.fillna(titanic.age.mean(),inplace=True)
```

```python
titanic
```

* 성별,선실별,출발지별 생존률

```python
#성별 생존률
titanic.groupby('sex')[['survived']].mean()
```

```python
#선실별 생존률
titanic.groupby('class')[['survived']].mean()
```

```python
titanic.pivot_table('survived','class')
```

```python
#출발지별 생존률
titanic.pivot_table('survived','embarked')
```

* 미성년자,청년,중년,장년,노년

```python
 bins=[1,20,30,50,70,100]
 labels=['미성년자','청년','중년','장년','노년']
 titanic['age_cat']=pd.cut(titanic.age,bins,labels=labels,include_lowest=True)
 titanic.tail(2)
```

```python
titanic.age_cat.value_counts()
```

```python
titanic.pivot_table('survived','age_cat')
```

* qcut을 활용하여 3개의 나이 그룹

```python
titanic['age_group']=pd.qcut(titanic.age,3,labels=['A1','A2','A3'])
titanic.tail()
```

```python
titanic.age_group.value_counts()
```

```python
titanic.groupby('age_group')[['survived']].mean()
```
