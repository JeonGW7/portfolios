* 실수값을 카테고리 값으로 변환

```python
ages=[1,2,20,23,37,31,61,20,41,32,100]
bins=[1,20,30,50,70,100]
labels=['미성년자','청년','중년','장년','노년']
cats=pd.cut(ages,bins,labels=labels)
cats
```

```python
print(cats.categories)
```

```python
print(cats.codes)
```

```python
adf=pd.DataFrame(ages,columns=['ages'])
adf['age_cat']=pd.cut(adf.ages,bins,labels=labels,include_lowest=True)
adf
```
