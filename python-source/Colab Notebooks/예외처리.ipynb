```python
f=open('test.txt','w')
try:
  data=f.read()
  print(data)
except Exception as e:
  print(e)
finally:
  f.close()
```

```python
try:
  a=[1,2]
  print(a[3])
except ZeroDivisionError:
  print('0으로 나눌수 없습니다.')
except IndexError:
  print("인덱싱 할 수 없습니다")
```
