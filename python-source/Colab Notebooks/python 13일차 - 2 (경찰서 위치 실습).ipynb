```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')   # 화면없이 실행
options.add_argument('--no-sandbox')
options.add_argument("--single-process")
options.add_argument("--disable-dev-shm-usage")
driver = webdriver.Chrome('chromedriver', options=options)
```

```python
with open('kakaoapikey.txt') as file:
  kakao_key=file.read()
len(kakao_key)
```

```python
url = 'https://minwon.police.go.kr/#customerCenter/plcstSearch'
driver.get(url)
time.sleep(2)
```

```python
driver.find_element_by_css_selector('#tab3_1').click()
```

```python
local=driver.find_element_by_css_selector('#indexedA').click() #지역
time.sleep(1)
```

```python
soup=BeautifulSoup(driver.page_source,'html.parser')
trs=soup.select('.title')
del trs[0],trs[-1]
trs
```

```python
police_list=trs
for i in range(len(police_list)):
  police_list[i]=police_list[i].get_text().strip()

police_list
```

```python
df=pd.DataFrame({'경찰서 주소':police_list})
df
```

```python
driver.find_element_by_xpath('//*[@id="pagination"]/div/a[1]').click()
time.sleep(1)
soup=BeautifulSoup(driver.page_source,'html.parser')
trs2=soup.select('.title')
del trs2[0],trs2[-1]
trs2
```

```python
police_list2=[]
for i in range(6):
  driver.find_element_by_xpath('//*[@id="pagination"]/div/a[9]').click()
  time.sleep(1)
  soup=BeautifulSoup(driver.page_source,'html.parser')
  trs=soup.select('.title')
  del trs[0],trs[-1]
  police_list2.append(trs)

police_list2[0]
```

```python
for i in range(len(police_list2)):
  police_list2[i]=police_list2[i].get_text().strip()
```

```python
df2=pd.DataFrame({'경찰서 주소':police_list2})
df2
```

```python
df=pd.DataFrame({'경찰서주소':police_list+police_list2})
df
```

```python
df.to_csv('서울경찰서.csv',index=False)
```

```python
import requests
from urllib.parse import quote
```

```python
local_url='https://dapi.kakao.com/v2/local/search/address.json'
addr='서울특별시 송파구 중대로 221'
url=f'{local_url}?query={quote(addr)}'
url
```

```python
header = {'Authorization': f'KakaoAK {kakao_key}'}
```

```python
df.경찰서주소[1]
```

```python
result=requests.get(url,headers=header).json()
result
```

```python
result['documents'][0]['x']
```

```python
len(df.index)
```

```python
lng_list,lat_list=[],[]
for i in df.index:
  url=f'{local_url}?query={quote(df.경찰서주소[i])}'
  result=requests.get(url,headers=header).json()
  try:
    lng_list.append(float(result['documents'][0]['x']))
    lat_list.append(float(result['documents'][0]['y']))
  except:
    print(df.경찰서주소[i])
```

```python
lng_list.insert(1,'-')
lng_list.insert(16,'-')
lat_list.insert(1,'-')
lat_list.insert(16,'-')
```

```python
df['위도']=lat_list
df['경도']=lng_list
df
```
