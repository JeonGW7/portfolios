```python
class 참치선물세트():
  일반=0
  야채=0
  고추=0
  def 총합(self,이름):
    합=self.일반+self.야채+self.고추
    return 이름+str(합)
  def 출력(self,이름):
    print('**',이름,'**')
    print('일반참치:',self.일반)
    print('야채참치:',self.야채)
    print('고추참치:',self.고추)
```

```python
참01호=참치선물세트()
참01호.일반=12
참01호.야채=3
참01호.고추=3
```

```python
출력값=참01호.총합('참치선물세트01호 내용물:')
print(출력값)
```

##생성자

```python
class 참치선물세트():
  일반=0
  야채=0
  고추=0
  def __init__(self,일반,야채,고추):
    self.일반=일반
    self.야채=야채
    self.고추=고추
```

```python
참01호=참치선물세트(3,2,3)
print(참01호)
```

```python
class 세면도구:
  def __init__(self,치약,샴푸,비누):
    self.치약=치약
    self.샴푸=샴푸
    self.비누=비누

  def 내용물출력(self,세트이름):
    print(세트이름)
    print('치약',self.치약)
    print('샴푸',self.샴푸)
    print('비누',self.비누)

class 특별세트(세면도구):
  def __init__(self,치약,샴푸,비누,면도기):
    super().__init__(치약,샴푸,비누)
    self.면도기=면도기
    print('면도기',self.면도기)

특01호=특별세트(1,2,3,1)
특01호.내용물출력("명절 특별세트 01호")
```

```python
class Calculator:
  def __init__ (self): 
    self.result = 0
  def add(self, num): 
    self.result += num 
    return self.result
```

```python
cal1 = Calculator() 
cal2 = Calculator()
```

```python
print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```

#파이썬 객체 저장/로드
