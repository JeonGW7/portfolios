* 그림범위

```python
plt.plot([0,1,2,3],[0,1,4,9],'ro-')
plt.title('x축 y축 범위지정')
plt.xlim(-1,4)  #line limit
plt.ylim(-1,10)
plt.grid()
plt.show()
```

* 곡선 그림

```python
X=np.arange(0,2*np.pi,0.01)
Ysin,Ycos=np.sin(X),np.cos(X)
plt.plot(X,Ysin)
plt.plot(X,Ycos)
plt.title('Sin,Cos 그래프')
plt.grid()
plt.show()
```

* Tick 설정

```python
plt.plot(X,Ysin)
plt.plot(X,Ycos)
plt.title('Sin,Cos 그래프')
plt.xticks([0,np.pi/2,np.pi,3*np.pi/2,2*np.pi],[r'$0$',r'$\pi/2$',r'$\pi$',r'$3\pi$/2',r'$2\pi$'])
plt.yticks([-1,0,1])
plt.grid()
plt.show()
```

```python
plt.plot(X,Ysin,label='sin')  #범례
plt.plot(X,Ycos,label='cos')
plt.title('Sin,Cos 그래프')
plt.xticks([0,np.pi/2,np.pi,3*np.pi/2,2*np.pi],[r'$0$',r'$\pi/2$',r'$\pi$',r'$3\pi$/2',r'$2\pi$'])
plt.yticks([-1,0,1])
plt.legend(loc=0)  #범례 위치
plt.grid()
plt.show()
```

* x축과 y축의 label

```python
plt.plot(X,Ysin,label='sin')  #범례
plt.plot(X,Ycos,label='cos')
plt.title('Sin,Cos 그래프',fontsize=15)
plt.xticks([0,np.pi/2,np.pi,3*np.pi/2,2*np.pi],[r'$0$',r'$\pi/2$',r'$\pi$',r'$3\pi$/2',r'$2\pi$'])
plt.yticks([-1,0,1])
plt.xlabel('시간')
plt.ylabel('진폭')
plt.legend(loc=0)  #범례 위치
plt.grid()
plt.show()
```

* 그림 크기

```python
plt.figure(figsize=(10,6))
plt.plot(X,Ysin,label='sin')  #범례
plt.plot(X,Ycos,label='cos')
plt.title('Sin,Cos 그래프',fontsize=15)
plt.xticks([0,np.pi/2,np.pi,3*np.pi/2,2*np.pi],[r'$0$',r'$\pi/2$',r'$\pi$',r'$3\pi$/2',r'$2\pi$'])
plt.yticks([-1,0,1])
plt.xlabel('시간')
plt.ylabel('진폭')
plt.legend(loc=0)  #범례 위치
plt.grid()
plt.show()
```
