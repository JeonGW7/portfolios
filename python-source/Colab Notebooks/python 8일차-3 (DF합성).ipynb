```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

* inner join

데이터가 있는 값만 들어간다

```python
df3= pd.merge(df1,df2)
df3
```

+ outer join

```python
#full outer join
df4=pd.merge(df1,df2,how='outer')
df4.head(8)
```

```python
pd.merge(df1,df2,how='left')  #df1을 기준으로 조인, right로 하면 df2를 기준으로 조인
```

* join method

```python
#인덱스가 join하는 역할
df1.set_index('고객번호',inplace=True)
df2.set_index('고객번호',inplace=True)
df1
```

```python
df2
```

```python
df2.join(df1)
```
